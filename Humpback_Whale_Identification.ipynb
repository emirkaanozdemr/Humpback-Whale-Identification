```python
import pandas as pd
ss=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Humpback Whale /sample_submission.csv")
train=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Humpback Whale /train.csv")
```

```python
print(train.head())
print(ss.head())
```

```python
train["Id"].value_counts()
```

```python
import cv2
import os

x = []
for img_path in train["Image"]:
    if os.path.isfile(img_path):
        img = cv2.imread(str(img_path))
        if img is not None:
            img = cv2.resize(img, (170, 170))
            img = img / 255.0
            x.append(img)
        else:
            print(f"Error: Unable to read image at path {img_path}")
    else:
        print(f"Error: File not found at path {img_path}")
```

```python
import numpy as np
x=np.array(x)
```

```python
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train["Id"] = le.fit_transform(train["Id"])
y = train["Id"]
```

```python
test=[]
for img_path in ss["Image"]:
        img = cv2.imread(str(img_path))
        img = cv2.resize(img, (170, 170))
        img = img / 255.0
        test.append(img)
```

```python
test=np.array(test)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, Dense, Flatten, Input, MaxPooling2D, Dropout, BatchNormalization
```

```python
print(y.max()-y.min()+1)
```

```python
model = Sequential()
model.add(Input(shape=(170, 170, 3)))
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(60, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(80, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(100, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(120, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128))
model.add(BatchNormalization())
model.add(Dense(5005, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=10, batch_size=32, validation_split=0.2)
```

```python
model.save("model.h5")
```

```python
predictions=model.predict(test)
```

```python
top_5_predictions = np.argsort(predictions, axis=1)[:, -5:][:, ::-1]
```

```python
ss["Id"]=top_5_predictions
```

```python
results = []
for idx, prediction in enumerate(top_5_predictions):
    result = {'Image': ss.Image[idx]}
    for i in range(5):
        result[f'Prediction_{i+1}'] = prediction[i]
    results.append(result)
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df.to_csv("submission.csv", index=False)
```
